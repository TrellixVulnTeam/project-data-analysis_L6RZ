```python
# Store the domestic and arrest columns in a dictionary 
# to use conditionals
# check where arrest and domestic are True
domestic_arrest = {
    'arrest': chicago_crime['arrest'],
    'domestic': chicago_crime['domestic']
}

type(domestic_arrest)
```

```python
# show where the arrest and domestic are True
for key, values in domestic_arrest['arrest'].items():
    if key and values == True:
        print(key, values)
```

```python
# show where the arrest and domestic are True
for key, values in domestic_arrest['domestic'].items():
    if key and values == True:
        print(key, values)
```

```python
# variable to store the index
# where domestic and arrest are True
domestic_arrest_index_true = []

# show the index where domestic and arrest are both True 
for key1, values1 in domestic_arrest['domestic'].items(): # Get the domestic index and value
    if key1 and values1 == True: # if True
        for key2, values2 in domestic_arrest['arrest'].items(): # # Get the arrest index and value
            if key2 and values2 == True: # if True
                if key1 == key2: # both arrest and domestic index must be the same 
                    domestic_arrest_index_true.append(key1)
                    print(key1) # index that meets the conditionals
print(domestic_arrest_index_true)
```

```python
# check the list length
# This is the total where the crime happends domestically and the assialant(s) is/are arrested
len(domestic_arrest_index_true)
```

```python
# validate the list
count_list = 0

for key1, values1 in domestic_arrest['domestic'].items(): # Get the domestic index and value
    if key1 and values1 == True: # if True
        for key2, values2 in domestic_arrest['arrest'].items(): # # Get the arrest index and value
            if key2 and values2 == True: # if True
                if key1 == key2: # both arrest and domestic index must be the same 
                    count_list += 1
                    print(key1, values1, key2, values2)
```

```python
# check
print(count_list)
```

Note: list and the check have the same length and values, thus we can use the list (index) for analysis

```python
# Turn the list into pandas Series
index_domestic_arrest = pd.Series(domestic_arrest_index_true)
index_domestic_arrest.head()
```
